This notebook demonstrates **the evolution of NLP techniques** â€” from **Rule-Based Systems** to **AI Agents**,  
using a real-world use case: deciding if a plasma donor should be **deferred** or **accepted** based on text input.

## ğŸ§© 1ï¸âƒ£ Rule-Based System

```python

text = "I had a fever last week but tested negative yesterday."

if "fever" in text.lower():
    decision = "Defer donor"
else:
    decision = "Accept donor"

print(decision)
```

## ğŸ”¤ 2ï¸âƒ£ Lemmatization / Normalization

```python

import spacy
nlp = spacy.load("en_core_web_sm")

text = "I felt feverish yesterday."
doc = nlp(text.lower())
lemmas = [token.lemma_ for token in doc]

print("Lemmatized:", lemmas)
if "fever" in lemmas:
    print("Possible fever case")
```

## ğŸ“Š 3ï¸âƒ£ Bag of Words (BoW)

```python

from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "I have a fever today",
    "I had a fever last week but fine now",
    "No fever today"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

## âš–ï¸ 4ï¸âƒ£ TF-IDF (Weighted Words)

```python

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

## ğŸ§  5ï¸âƒ£ Word2Vec (Semantic Meaning)

```python

from gensim.models import Word2Vec

sentences = [
    ["i", "had", "a", "fever", "last", "week"],
    ["i", "felt", "feverish", "yesterday"],
    ["no", "fever", "today"]
]

model = Word2Vec(sentences, vector_size=50, min_count=1, window=3)
print(model.wv.most_similar("fever"))
```

## ğŸ” 6ï¸âƒ£ Recurrent Neural Network (RNN)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

texts = [
    "I have a fever today",
    "I had a fever last week but fine now",
    "No fever today"
]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = pad_sequences(tokenizer.texts_to_sequences(texts))

model = Sequential([
    Embedding(len(tokenizer.word_index)+1, 8),
    SimpleRNN(16, activation='tanh'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

## â³ 7ï¸âƒ£ LSTM (Long Short-Term Memory)

```python
from tensorflow.keras.layers import LSTM

model = Sequential([
    Embedding(len(tokenizer.word_index)+1, 8),
    LSTM(32),
    Dense(1, activation='sigmoid')
])
model.summary()
```

## âš¡ 8ï¸âƒ£ Transformer (Self-Attention)

```python
from transformers import pipeline
import torch

# Use PyTorch backend to avoid TensorFlow/Keras compatibility issues
classifier = pipeline("text-classification", 
                     model="distilbert-base-uncased-finetuned-sst-2-english",
                     framework="pt")  # Explicitly use PyTorch
print(classifier("I had a fever last week but tested negative yesterday"))
```

## ğŸ” 9ï¸âƒ£ Retrieval-Augmented Generation (RAG)

```python
# RAG implementation using simpler approach due to complexity of full RAG setup
from transformers import pipeline

# Use a question-answering model as a simplified RAG demonstration
qa_pipeline = pipeline("question-answering", framework="pt")

# Simulate a knowledge base (in real RAG, this would be retrieved from a vector database)
context = """
Plasma donation deferral policies:
- Donors with fever should be deferred for 14 days after symptoms resolve.
- Donors who tested negative after illness can donate 7 days post-recovery.
- Current illness or active symptoms require immediate deferral.
- Temperature above 99.5Â°F (37.5Â°C) results in automatic deferral.
"""

question = "What is the deferral policy for fever last week?"

# Use the QA model to extract answer from context
result = qa_pipeline(question=question, context=context)
print(f"Question: {question}")
print(f"Answer: {result['answer']}")
print(f"Confidence: {result['score']:.3f}")
```

## ğŸ¤– ğŸ”Ÿ AI Agent (Multi-Step Workflow)

```python

def retrieve_policy(query):
    return "Donor should be deferred for 14 days after last fever."

def summarize_donor_input(text):
    return f"Summary: {text}"

def draft_decision(summary, policy):
    return f"Decision based on {summary}: {policy}"

text = "I had a fever last week but fine today."
summary = summarize_donor_input(text)
policy = retrieve_policy(text)
final_note = draft_decision(summary, policy)

print(final_note)
```

## âœ… Summary

| Stage | Key Idea | Limitation | Next Step |
|--------|-----------|-------------|------------|
| Rule-Based | Hardcoded rules | No context | Lemmatization |
| Lemmatization | Word roots | No meaning | BoW |
| BoW/TF-IDF | Word counts | No semantics | Word2Vec |
| Word2Vec | Semantic vectors | No time context | RNN |
| RNN | Sequential memory | Forgets long-term | LSTM |
| LSTM | Long-term memory | Slow | Transformer |
| Transformer | Attention/context | Static data | RAG |
| RAG | Real-time retrieval | Single model | AI Agents |
| AI Agents | Multi-role orchestration | â€” | â€” |
